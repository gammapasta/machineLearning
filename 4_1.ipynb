분류문제에서 클래스의 확률을 예측  
5개 특성 기반으로 7개클래스 분류  

```python
import pandas as pd
fish = pd.read_csv('http://bit.ly/fish_csv_data')
fish.head()
```

```python
print(fish.shape)
```

```python
#어떤 종류의 클래스가 있는지 나열
print(pd.unique(fish['Species']))
```

```python
타켓 만들기 데이터 프레임에서 종족 열을 탓[ㄱ으로 지정, 나머지 5개의 열은 입력 데이터로 지정 괄호 2개는 2차원 배열!!!!!!!!!!!]
```

```python
fish_input = fish[['Weight','Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
#밑은 1차원

print(fish_input[:5])
```

```python
#정규화
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

과소적합, 분류 문제에서 socre 반환하는 값은 정확도이며 회귀문제에서는 결정계수를 반환
예측하기

```python
print(kn.predict(test_scaled[:5]))
#5번쨰까지 그런데 결과가 0과1이 아니라(2진분류) 다중 분류라
```

#다중분류  
3개 이상 클래스 포함  


```python
#이제 확률 계산하기
import numpy as np

proba = kn.predict_proba(test_scaled[:5]) #클래스별 확률값 반환
print(np.round(proba, decimals=4)) #소수점 네번쨰 자리까지 반올림
```

```python
print(kn.classes_)
```

```python
#5번쨰 샘플의 최근접 이웃의 클래스 확인

distances, indexes = kn.kneighbors(test_scaled[[4]])   #5번쨰값은 2차원이라 괄호 2개
distances, indexes = kn.kneighbors(test_scaled[4:5])
print(train_target[indexes])
```

```python
#3개의 최근접 이용시 가능한 확률은 0 0.333 0.666 1
#그렇다면 n_neightebo를 3이 아니라 더 큰 값으로 바꾸면
#확률의 다양성이 증가
#훈련 세트와 테스트 세트의 점수가 높은 최대 n_neighbers 값을 찾아보자

kn = KNeighborsClassifier(n_neighbors=10)
kn.fit(train_scaled, train_target)

print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
i = 1
while i<10:
  kn = KNeighborsClassifier(n_neighbors=i)
  kn.fit(train_scaled, train_target)
  print(i,'번쨰 ')
  print(kn.score(train_scaled, train_target))
  print(kn.score(test_scaled, test_target))
  print(" ")
  i = i + 1
#음... 3이 가장 높믐
```

#로지스틱 회기 (분류임)  
z = a무게 + b갈아 + c대각선 + d높이 + e두깨 + f  

문제는 z값은 어떤 값도 가능함  
확률이 되려면 0부터 1사이의 값이 되어야함  
z가 아주 큰 음수일때 0 큰 양수일떄 1로 변환하는 시그모이드 함수 사용

```python
#이건 그냥 시그모이드 함수

import numpy as np
import matplotlib.pyplot as plt

z = np.arange(-5, 5, 0.1)
phi = 1 / (1 + np.exp(-z))

plt.plot(z, phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
```

```python
도미와 빙어 데이터만 이용해 로지스틱 회귀

7개의 클래스에서 2개의 데이터만 선택

불ㄹ;언 인덱싱 사ㅇ용
true false 값을 전달하여 행을 선택
```

```python
#boolean indexing
char_arr = np.array(['A', 'B', 'C', 'D', 'E'])
print(char_arr[[True, False, True, False, False]])
```

```python
#여기서 함수 쓰는거 없어서 외어야함
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt') #bream or smelt
print(bream_smelt_indexes[:])
```

```python
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
print(target_bream_smelt)
```

LogisticRegression 이거 이용  
- 훈련  
- 예측  
- 확률  
해보기

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
print(lr.predict(train_bream_smelt[:5]))
print(lr.predict_proba(train_bream_smelt[:5]))

#음성 양성 구분
print(lr.classes_)
#bream이 음성
```

```python
#공식에 사용되는 숫자들!!!! 이렇게 해서 z를 구함
print(lr.coef_, lr.intercept_)
```

```python
#z는 확률이 아니라서 확률로변환
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
#실수라 확률로 계산하기
```

```python
#양성 클래스에 대한 확룰 계산
from scipy.special import expit

print(expit(decisions))
#z에 대한 확률값이 나옴
#아니면 직접 구현해도 된다
```

```python
lr = LogisticRegression(C = 20, max_iter=1000) #매개변수 넣기

lr.fit(train_scaled, train_target) #객체생성, 훈련

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))

print(lr.predict(test_scaled[:5])) #예측

proba = lr.predict_proba(test_scaled[:5]) #예측한 확률을 숫자로 출력
print(np.round(proba, decimals=3))

print(lr.classes_)
```

```python
print(lr.coef_.shape, lr.intercept_.shape) #클래스마다 계수와 절편을 만든다 5개의 특성을 사용하므로 열은 5개 타깃은 7개이므로 7개

decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=2))
```

#실습

모델 파라미터:기울기, 계수   
하이퍼 파라미터: 직접 지정해야하는 값 (최적화의미)  
-> 어떤것이 하이퍼 파라미터인지 알고 ,어떻게 지정할까
- 패널티 값 수정

```python
c = 1
i = 100
para = 0
arr = [[],[],[]]
while c < 30:
  while i < 10000:
    lr = LogisticRegression(penalty='l2', C = c, max_iter=i, solver='lbfgs', )
    lr.fit(train_scaled, train_target)
    print(lr.score(train_scaled, train_target))
    print(lr.score(test_scaled, test_target))
    arr[0].append(para)
    arr[1].append(lr.score(train_scaled, train_target))
    arr[2].append(lr.score(test_scaled, test_target))
    para = para + 1
    i = i + 500
    c = c + 1
  i = 500

lr = LogisticRegression(penalty='l2', C = 20, max_iter=1000, solver='lbfgs', )
lr.fit(train_scaled, train_target)

print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
plt.plot(arr[0], arr[1])

plt.xlabel('c')
plt.ylabel('arr')
plt.show()
```
